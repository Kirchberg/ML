## 4. Отобразите данные по индексу CPI за 2020 год в виде горизонтального столбчатого графика

```python
# 4. Отобразите данные по индексу CPI за 2020 год в виде горизонтального столбчатого графика

raw_countries = xl_file[[xl_file.columns[0]]].to_numpy().flatten()
raw_cpi_score = xl_file[[xl_file.columns[3]]].to_numpy().flatten()

countries = raw_countries[2:]
cpi_scores = raw_cpi_score[2:]

# Сделал только первые 15, чтобы все влезло
plt.style.use('seaborn-whitegrid')
df = pd.DataFrame({'Score': cpi_scores[:15]}, columns=['Score'], index = countries[:15])
df.plot.barh(figsize=(10, 4)).invert_yaxis()
plt.show()
```
