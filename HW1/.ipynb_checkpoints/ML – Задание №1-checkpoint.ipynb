## 4. Отобразите данные по индексу CPI за 2020 год в виде горизонтального столбчатого графика

```python
# 4. Отобразите данные по индексу CPI за 2020 год в виде горизонтального столбчатого графика

raw_countries = xl_file[[xl_file.columns[0]]].to_numpy().flatten()
raw_cpi_score = xl_file[[xl_file.columns[3]]].to_numpy().flatten()

countries = raw_countries[2:]
cpi_scores = raw_cpi_score[2:]

# Сделал только первые 15, чтобы все влезло
plt.style.use('seaborn-whitegrid')
df = pd.DataFrame({'Score': cpi_scores[:15]}, columns=['Score'], index = countries[:15])
df.plot.barh(figsize=(10, 4)).invert_yaxis()
plt.show()
```

## 7. Постройте графики индекса CPI за 2012-2020 для стран своего региона

```python
df_eca_cpi_score = new_df_group[[new_df_group.columns[0], new_df_group.columns[3], new_df_group.columns[7], new_df_group.columns[11], new_df_group.columns[15], new_df_group.columns[19], new_df_group.columns[22], new_df_group.columns[25], new_df_group.columns[28], new_df_group.columns[31]]]
df_eca_cpi_score
```

```python
array_names = ["CPI Score 2012","CPI Score 2013","CPI Score 2014","CPI Score 2015","CPI Score 2016","CPI Score 2017","CPI Score 2018","CPI Score 2019","CPI Score 2020"]
mean_array = []
array = []
russia_array = []
russia_index = df_eca_cpi_score[df_eca_cpi_score.columns[0]].loc[lambda x: x=="Russia"].index
print(russia_index)
for col_name in df_eca_cpi_score.columns:
    if "Unnamed" in col_name:
        array.append(df_eca_cpi_score[col_name].values)
        mean_array.append(df_eca_cpi_score[col_name].values.mean())
        russia_array.append(df_eca_cpi_score[col_name].values[-4])
plt.figure(figsize=(7,7))
plt.plot(array, 'o-', color='black')
plt.ylabel("CPI")
plt.xticks(range(len(array)),array_names, rotation='vertical')
plt.show()
```
